```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# settings
sin = np.sin
cos = np.cos
pi = np.pi
sqrt = np.sqrt
exp = np.exp
arcsin = np.arcsin
arccos = np.arccos
e = np.e
```

# Vector field

```python
def plot_vec_field(u, v, lim=8, size=12, streamplot=False):
    
    """
    """
    
    # create a cartesian mesh grid
    n = int(2*lim) + 1    # number of points
    x = np.linspace(-lim, lim, n)
    x, y = np.meshgrid(x, x)
    
        
    # creat a polar meshgrid from cartesian using known transformations
    r = np.sqrt(x**2 + y**2)    # r as function of (x, y)
    theta = np.arctan2(y, x)    # theta in function of (x, y)

        
    # eval
    u = eval(u)
    v = eval(v)
    
    # calculate vector size at each point and use it as color
    color = np.sqrt(u**2 + v**2)
    
    # plot
    plt.figure(figsize=(size+2,size))
    
    if streamplot:
        plt.streamplot(x, y, u, v, color=color)
        plt.title("Stream plot")
    else:
        plt.title("Quiver plot")
        plt.quiver(x, y, u, v, color)
    
    
    # layout
    plt.axhline(c="k")
    plt.axvline(c="k")
    plt.colorbar()
    plt.grid()
    plt.show()
```

```python
interact(plot_vec_field, u="y", v="-x+y", lim=(0.0,20.0,0.01), size=(5,20))
```
